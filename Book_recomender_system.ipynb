```python
x=ratings["User-ID"].value_counts()>200
```

```python
ratings=ratings[ratings["User-ID"].isin(y)]
```

```python
books=pd.read_csv("/content/BX-Books.csv", sep=";", on_bad_lines='skip', encoding="latin1", engine='python')
books.columns = ['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('Book-Title')['Book-Rating'].count().reset_index().rename(columns={"Book-Rating": "number_of_rating"})
```

```python
number_rating.sample(5)
```

```python
final_rating=ratings_with_books.merge(number_rating, on="Book-Title")
```

```python
final_rating.sample(4)
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating["number_of_rating"]>=40]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(["User-ID","Book-Title"],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns="User-ID", index="Book-Title", values="Book-Rating")
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
!pip install scipy
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from  sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.index==237
```

```python
distances, suggestions = model.kneighbors(book_sparse[book_pivot.index.get_loc('The Poisonwood Bible: A Novel')].reshape(1,-1), n_neighbors=6)

print("Books similar to 'The Poisonwood Bible: A Novel':")
for i in range(len(suggestions[0])):
    if i != 0:
        print(book_pivot.index[suggestions[0][i]])
```

```python
def book_recomendation(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances, suggestions = model.kneighbors(book_sparse[book_pivot.index.get_loc(book_name)].reshape(1,-1), n_neighbors=6)

  print("Books similar to", book_name, ":")
  for i in range(len(suggestions[0])):
    if i != 0:
        print(book_pivot.index[suggestions[0][i]])
```

```python
book_recomendation("The Lovely Bones: A Novel")
```
